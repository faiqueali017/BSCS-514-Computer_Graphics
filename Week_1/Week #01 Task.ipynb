### Reading data from directory

```python
import os

folder_path = r'C:\Users\Rabia\Desktop\BSCS-514-Computer_Graphics\Week_1\SampleData'

#Read directory data and display the filenames and its path
def listDir(dir):
    fileNames = os.listdir(dir)
    for fileName in fileNames:
        print('File Name: '+fileName)
        print('Folder Path: '+os.path.abspath(os.path.join(dir, fileName)), sep='\n')
        
        
if __name__ == '__main__':
    listDir(folder_path)
```

```python
#Count total no of files in a directory
len(os.listdir(folder_path))
```

### Count only particular type of file/data

```python
import glob

tifCounter = len(glob.glob1(folder_path,"*.jpg"))
print(tifCounter)
```

### Calculate Total size and total count of a particular data in a directory

```python
def calculateFilesSizes(folder_path, file_extension):
    files = glob.glob1(folder_path,file_extension)
    total_size = 0
    
    for file in files:
        total_size += os.path.getsize(os.path.join(folder_path, file))
    print("No of files:",len(files))
    print("Total file size:",total_size,"bytes")

#For sample testing we are calculating for .mp3 and .jpg
calculateFilesSizes(folder_path, "*.mp3")
calculateFilesSizes(folder_path, "*.jpg")
```

### Histogram plotting of all file sizes to its frequency

```python
import matplotlib.pyplot as plt

#Per file size in mbs
file_sizes = [0.2,0.5,0.2,3,5,5,1,1,1,10,12,12,12,23,22,80,43,46,42,40,56,10,5,3,2,66,1,1.2,0.2,0.7,3,3.5,63,63.5]
file_size_range = [0,10,20,30,40,50,60,70,80]

#Total sample files data count
print("Total sample files count:",len(file_sizes))

plt.hist(file_sizes,file_size_range,rwidth=0.9)
plt.xlabel('Size of files in MB')
plt.ylabel('Frequency')
plt.yticks(range(0,20,2))
plt.show()
```

### Bar graph to represent different types of device data and its total count.

```python
fig = plt.figure(figsize=(7,5))

#Sample device data
names = ['music','docs','videos','photos','others']
count = [230,60,43,520,75]
position = [0,1,2,3,4]
colors = ['red','green','purple','orange','gray']

plt.bar(position, count, width = 0.8, color = colors)
plt.xlabel('Types of data')
plt.ylabel('Count')
plt.xticks(position,names)

plt.show()
```

### Bar graph to represent different types of device data and its total size.

```python
fig = plt.figure(figsize=(7,5))

#Sample device data
data_names = ['music','docs','videos','photos','others']
data_size = [1150,180,5160,2600,1875]
position = [0,1,2,3,4]
colors = ['red','green','purple','orange','gray']

plt.bar(position, data_size, width = 0.8, color = colors)
plt.xlabel('Types of data')
plt.ylabel('Data Size in MB')
plt.xticks(position,data_names)

plt.show()
```
